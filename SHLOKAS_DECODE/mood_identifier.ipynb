```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_excel("Dataset.xlsx")
```

```python
df
```

```python
text = input("Enter text : ")
```

```python
import nltk
from nltk.tokenize import word_tokenize
text = word_tokenize(text)
```

```python
from nltk.stem import SnowballStemmer
stemming = SnowballStemmer("english")
length = len(text)

for i in range(0,length):
    text[i] = stemming.stem(text[i])
print(text)
```

```python
# Compare input text with each mood list
counts = {}

for i in range(len(df)):
    mood = df.iloc[i, 0]
    mood_words = df.iloc[i, 1]
    count = sum(1 for word in text if word in mood_words)
    counts[mood] = count

# Display results
print("Common word counts per mood:")
for mood, c in counts.items():
    print(f"{mood}: {c}")

# Detect mood with highest match
detected_mood = max(counts, key=counts.get)
print("\nDetected Mood:", detected_mood)
```
